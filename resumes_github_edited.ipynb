```python
import random
import math


def hierarchy_pos(G, root=None, width=1., vert_gap = 0.2, vert_loc = 0, xcenter = 0.5):

    '''
    From Joel's answer at https://stackoverflow.com/a/29597209/2966723.
    Licensed under Creative Commons Attribution-Share Alike

    If the graph is a tree this will return the positions to plot this in a
    hierarchical layout.

    G: the graph (must be a tree)

    root: the root node of current branch
    - if the tree is directed and this is not given,
      the root will be found and used
    - if the tree is directed and this is given, then
      the positions will be just for the descendants of this node.
    - if the tree is undirected and not given,
      then a random choice will be used.

    width: horizontal space allocated for this branch - avoids overlap with other branches

    vert_gap: gap between levels of hierarchy

    vert_loc: vertical location of root

    xcenter: horizontal location of root
    '''
    if not nx.is_tree(G):
        raise TypeError('cannot use hierarchy_pos on a graph that is not a tree')

    if root is None:
        if isinstance(G, nx.DiGraph):
            root = next(iter(nx.topological_sort(G)))  #allows back compatibility with nx version 1.11
        else:
            root = random.choice(list(G.nodes))

    def _hierarchy_pos(G, root, width=1., vert_gap = 0.2, vert_loc = 0, xcenter = 0.5, pos = None, parent = None):
        '''
        see hierarchy_pos docstring for most arguments

        pos: a dict saying where all nodes go if they have been assigned
        parent: parent of this branch. - only affects it if non-directed

        '''

        if pos is None:
            pos = {root:(xcenter,vert_loc)}
        else:
            pos[root] = (xcenter, vert_loc)
        children = list(G.neighbors(root))
        if not isinstance(G, nx.DiGraph) and parent is not None:
            children.remove(parent)
        if len(children)!=0:
            dx = width/len(children)
            nextx = xcenter - width/2 - dx/2
            for child in children:
                nextx += dx
                pos = _hierarchy_pos(G,child, width = dx, vert_gap = vert_gap,
                                    vert_loc = vert_loc-vert_gap, xcenter=nextx,
                                    pos=pos, parent = root)
        return pos


    return _hierarchy_pos(G, root, width, vert_gap, vert_loc, xcenter)
```

```python
class Graph2Vec(Estimator):
    """An implementation of "graph2vec" (https://arxiv.org/abs/1707.05005).
    It generates representations for graphs using Weisfeiler-Lehman tree features."""

    def __init__(
        self,
        wl_iterations: int = 2,
        attributed: bool = False,
        dimensions: int = 512,
        workers: int = 4,
        down_sampling: float = 0.0001,
        epochs: int = 10,
        learning_rate: float = 0.025,
        min_count: int = 5,
        seed: int = 42,
        erase_base_features: bool = False,
    ):
        self.wl_iterations = wl_iterations
        self.attributed = attributed
        self.dimensions = dimensions
        self.workers = workers
        self.down_sampling = down_sampling
        self.epochs = epochs
        self.learning_rate = learning_rate
        self.min_count = min_count
        self.seed = seed
        self.erase_base_features = erase_base_features

    def fit(self, graphs: List[nx.Graph]):
        self._set_seed()

        documents = [WeisfeilerLehmanHashing(graph, self.wl_iterations, self.attributed, self.erase_base_features) for graph in graphs]
        documents = [TaggedDocument(words=doc.get_graph_features(), tags=[str(i)]) for i, doc in enumerate(documents)]

        self.model = Doc2Vec(
            documents,
            vector_size=self.dimensions,
            window=0,
            min_count=self.min_count,
            dm=0,
            sample=self.down_sampling,
            workers=self.workers,
            epochs=self.epochs,
            alpha=self.learning_rate,
            seed=self.seed,
        )

        self._embedding = [self.model.docvecs[str(i)] for i, _ in enumerate(documents)]

    def get_embedding(self) -> np.array:
        """Getting the embedding of graphs."""
        return np.array(self._embedding)

    def infer(self, graphs) -> np.array:
        """Infer the graph embeddings."""
        self._set_seed()

        documents = [WeisfeilerLehmanHashing(graph, self.wl_iterations, self.attributed, self.erase_base_features) for graph in graphs]
        documents = [doc.get_graph_features() for _, doc in enumerate(documents)]

        embedding = np.array(
            [self.model.infer_vector(doc, alpha=self.learning_rate, min_alpha=0.00001, epochs=self.epochs).astype('float64') for doc in documents]
        )

        return embedding

# Create an instance of the Graph2Vec class
model = Graph2Vec()

# Train the model on graphs
model.fit(graphs)

# Get embeddings for graphs
embeddings = model.infer(graphs)
```

```python
from dgl.data import DGLDataset

class SyntheticDataset(DGLDataset):
    def __init__(self):
        super().__init__(name="synthetic")

    def process(self):
        self.graphs = []
        self.labels = []
        with open('/content/drive/MyDrive/Colab Notebooks/graph_data_list_new.pickle', 'rb') as handle:
          graph_list = pickle.load(handle)

        with open('/content/drive/MyDrive/Colab Notebooks/new_salary.pickle', 'rb') as handle:
          label_list = pickle.load(handle)

        # For each graph ID...
        for graph_id in range(len(graph_list)):
            num_nodes = len(graph_list[graph_id].nodes())
            label = label_list[graph_id]

            # Create a graph and add it to the list of graphs and labels.
            g = dgl.from_networkx(graph_list[graph_id])
            central_node_features = []
            for key, value in dict(graph_list[graph_id].nodes()).items():
                  if key != 'salary':
                            central_node_features.append(value['label'])

            # Создаем features для центрального узла
            x = torch.tensor([central_node_features], dtype=torch.float)[0]

            g.ndata['feat'] = x
            g = dgl.add_self_loop(g)
            self.graphs.append(g)
            self.labels.append(label)

        # Convert the label list to tensor for saving.
        self.labels = torch.LongTensor(self.labels)

    def __getitem__(self, i):
        return self.graphs[i], self.labels[i]

    def __len__(self):
        return len(self.graphs)


dataset = SyntheticDataset()
graph, label = dataset[0]
print(graph, label)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from dgl.nn import GATConv

class GAT(nn.Module):
    def __init__(self, in_feats, hid_feats, out_feats, num_heads, dropout):
        super(GAT, self).__init__()
        self.conv1 = GATConv(in_feats, hid_feats, num_heads, dropout, dropout)
        self.conv2 = GATConv(hid_feats * num_heads, 32, num_heads, dropout, dropout)
        self.conv3 = GATConv(32 * num_heads, 16, num_heads, dropout, dropout)
        self.conv4 = GATConv(16 * num_heads, out_feats, 1, dropout, dropout)  # Последний слой с 1 головой
        self.fc = nn.Linear(out_feats, 1)  # Добавляем полносвязный слой для регрессии

    def forward(self, g, in_feats):
        h = in_feats
        h = self.conv1(g, h)
        h = F.elu(h).flatten(1)
        h = self.conv2(g, h)
        h = F.elu(h).flatten(1)
        h = self.conv3(g, h)
        h = F.elu(h).flatten(1)
        h = self.conv4(g, h)
        h = h.squeeze(1)


        # Агрегация репрезентаций узлов в репрезентацию графа
        g.ndata['h'] = h
        hg = dgl.mean_nodes(g, 'h')  # Усредняем репрезентации узлов

        # Получение конечного предсказания
        return self.fc(hg)  # Применяем полносвязный слой
```

```python
# Параметры модели
in_feats = 512  # Размерность входных признаков (из описания датасета)
hid_feats = 64  # Размерность скрытого слоя
out_feats = 1  # Размерность выходного слоя для регрессии
num_heads = 8  # Количество голов внимания
dropout = 0.6

# Создание модели
model = GAT(in_feats, hid_feats, out_feats, num_heads, dropout)
```

### LR 0.002

```python
# Функция потерь и оптимизатор
criterion = nn.MSELoss()  # MSE для задачи регрессии
optimizer = torch.optim.Adam(model.parameters(), lr=0.002)

# Обучение модели
epochs = 20

for epoch in range(epochs):
    for g, label in train_dataset:
        features = g.ndata['feat']  # Получаем признаки узлов
        label = label.float()  # Преобразуем метки в float

        # Проход вперед
        outputs = model(g, features)

        # Вычисление потери
        loss = criterion(outputs, label.reshape((1, 1)))

        # Обратный проход и шаг оптимизатора
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item()}')
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from dgl.nn import GATConv

class GAT(nn.Module):
    def __init__(self, in_feats, hid_feats, out_feats, num_heads, dropout):
        super(GAT, self).__init__()
        self.conv1 = GATConv(in_feats, hid_feats, num_heads, dropout, dropout)
        self.conv2 = GATConv(hid_feats * num_heads, out_feats, 1, dropout, dropout) # Последний слой с 1 головой
        self.fc = nn.Linear(out_feats, 1)  # Добавляем полносвязный слой для регрессии

    def forward(self, g, in_feats):
        h = in_feats
        h = self.conv1(g, h)
        h = F.elu(h).flatten(1)
        h = self.conv2(g, h)
        h = h.squeeze(1)


        # Агрегация репрезентаций узлов в репрезентацию графа
        g.ndata['h'] = h
        hg = dgl.mean_nodes(g, 'h')  # Усредняем репрезентации узлов

        # Получение конечного предсказания
        return self.fc(hg)  # Применяем полносвязный слой
```

```python
# Параметры модели
in_feats = 512  # Размерность входных признаков (из описания датасета)
hid_feats = 64  # Размерность скрытого слоя
out_feats = 1  # Размерность выходного слоя для регрессии
num_heads = 8  # Количество голов внимания
dropout = 0.6

# Создание модели
model = GAT(in_feats, hid_feats, out_feats, num_heads, dropout)
```

```python
# Функция потерь и оптимизатор
criterion = nn.MSELoss()  # MSE для задачи регрессии
optimizer = torch.optim.Adam(model.parameters(), lr=0.005)

# Обучение модели
epochs = 20

for epoch in range(epochs):
    for g, label in train_dataset:
        features = g.ndata['feat']  # Получаем признаки узлов
        label = label.float()  # Преобразуем метки в float

        # Проход вперед
        outputs = model(g, features)

        # Вычисление потери
        loss = criterion(outputs, label.reshape((1, 1)))

        # Обратный проход и шаг оптимизатора
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item()}')
```

```python
from datetime import datetime

# Функция для расчета возраста
def calculate_age(birth_date):
    # Преобразование строки в объект datetime
    birth_date = datetime.strptime(str(birth_date), '%Y-%m-%d %H:%M:%S')

    # Получение текущего времени
    today = datetime.now()

    # Разница в годах
    age = today.year - birth_date.year

    # Если день рождения еще не прошел в этом году, уменьшаем возраст на 1
    if today.month < birth_date.month or (today.month == birth_date.month and today.day < birth_date.day):
        age -= 1

    return age


birtday = df[df['birth_date'].notna()]['birth_date'].apply(calculate_age)

plt.figure(figsize=(12,4))
#df['birth_date'].value_counts().plot(kind='bar')
birtday.value_counts().plot(kind='bar')
plt.title('Возраст')
plt.xlabel('Возраст')
plt.ylabel('Количество записей')
plt.show()
```
